```python
x = torch.tensor([1.,2.,3.,4.])
y = torch.tensor([2.,4.,6.,8.])

m = torch.tensor(0.0)
c = torch.tensor(0.0)

y_pred = m * x + c
loss = ((y_pred - y) ** 2).mean()

print("Predicted Y : ", y_pred)
print("Loss : ",loss)
```
