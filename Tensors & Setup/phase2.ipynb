```python
import torch
import torch.nn as nn

# nn.Linear creates a fully connected / dense layer.
# Here:
# input_features = 3
# output_features = 2
#
# This means the layer expects 3 input values
# and produces 2 output values.
layer = nn.Linear(3, 2)

# Input tensor with one sample and three features.
#
# Shape of x = [1, 3]
# 1 means one data example
# 3 means three input features
#
# Example:
# [1.0, 2.0, 3.0] could represent [height, weight, age]
x = torch.tensor([[1.0, 2.0, 3.0]])

# Forward pass:
# The input x is passed through the linear layer.
#
# Formula:
# output = input * weights + bias
#
# Since this layer has 2 output features,
# y will contain 2 output values.
y = layer(x)

# In manual gradient descent, we update weights like this:
# w -= lr * w.grad
#
# But in real PyTorch models, we usually use optimizers.

# Example optimizer using SGD:
# torch.optim.SGD(model.parameters(), lr=0.01)

# Example optimizer using Adam:
# torch.optim.Adam(model.parameters(), lr=0.001)

# Standard PyTorch training steps:
#
# 1. optimizer.zero_grad()
#    Clears old gradients because PyTorch accumulates gradients by default.
#
# 2. Forward pass
#    Pass input through the model to get predictions.
#
# 3. Calculate loss
#    Compare prediction with the correct answer.
#
# 4. loss.backward()
#    Calculates gradients using backpropagation.
#
# 5. optimizer.step()
#    Updates model parameters using the gradients.

# Print the output of the linear layer.
print(y)

# General syntax:
# nn.Linear(input_features, output_features)

# Example:
# nn.Linear(3, 1)
#
# This means:
# Input: 3 features
# Output: 1 value
#
# Example use case:
# [height, weight, age] -> risk_score

# Another example:
# nn.Linear(3, 2)
#
# Input:
# 3 features
#
# Output:
# 2 values
#
# Example:
# [height, weight, age] -> [risk_score, health_score]
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Create a custom model class.
# Every PyTorch model should inherit from nn.Module.
class MyModel(nn.Module):
    def __init__(self):
        # Initialize the parent nn.Module class.
        # This is required so PyTorch can properly track layers and parameters.
        super().__init__()

        # Create a linear layer.
        # nn.Linear(1, 1) means:
        # input features = 1
        # output features = 1
        #
        # Formula:
        # y = weight * x + bias
        self.Linear = nn.Linear(1, 1)

    def forward(self, x):
        # Define the forward pass.
        # Whenever we call model(x), this function runs.
        return self.Linear(x)


# Training input data.
# Shape = [4, 1]
# There are 4 samples, and each sample has 1 feature.
x = torch.tensor([[1.], [2.], [3.], [4.]])

# Correct output values.
# The model should learn the relation:
# y = 2x
y = torch.tensor([[2.], [4.], [6.], [8.]])

# Create an object of the model.
model = MyModel()

# Define the loss function.
# MSELoss = Mean Squared Error Loss.
# It measures the average squared difference between predicted and actual values.
loss_fn = nn.MSELoss()

# Define the optimizer.
# Adam updates the model parameters using gradients.
# model.parameters() gives all learnable weights and biases of the model.
# lr = learning rate, controls the size of each update step.
optimiser = optim.Adam(model.parameters(), lr=0.1)

# Train the model for 10 epochs.
# One epoch means one complete pass through the training data.
for epoch in range(10):

    # Clear old gradients.
    # PyTorch accumulates gradients by default,
    # so we must reset them before each training step.
    optimiser.zero_grad()

    # Forward pass:
    # Give input x to the model and get predicted output.
    y_pred = model(x)

    # Calculate loss:
    # Compare predicted output with the correct output.
    loss = loss_fn(y_pred, y)

    # Backward pass:
    # Calculate gradients of the loss with respect to model parameters.
    loss.backward()

    # Update model parameters:
    # Optimizer updates weight and bias using calculated gradients.
    optimiser.step()

    # Print training progress.
    print(f"Epoch {epoch + 1}, Loss: {loss.item():.4f}")


# Save the trained model parameters.
# state_dict() contains only the learned weights and biases,
# not the full model class structure.
torch.save(model.state_dict(), "linear_model.pth")
```
