```python
import torch

# requires_grad = True
# torch.tensor(value, requires_grad = True)

w = torch.tensor([2.],requires_grad=True)  # this value will be tracked for learning purpose so that it can be altered during training phase to improve the model
x = torch.tensor(4)
y = w * x
print(y)    # tensor([8.], grad_fn=<MulBackward0>) , grad_fn knows how y is calculated
```

```python
x = torch.tensor(2.0)
w = torch.tensor(3.0, requires_grad=True)

y_pred = w * x
y_true = torch.tensor(10.0)

loss = ((y_pred - y_true) ** 2)

print(loss)
loss.backward()   # this will trace back and check who is responsible for the erorr and will change its value to reduce the error , in other words , it says how much each vulnerable number cause the mistake

print(w.grad)   # w.grad stores the instructions on how to change a number to reduce the mistake , first thing is it tells you the direction to change and the second thing is the amount 
# if w.grad > 0 then it will decrease and if w.grad < 0 then it will increase

print("Prediction : ",y_pred.item())
print("Loss : ",loss.item())
print("w_grad : ",w.grad.item())
```
