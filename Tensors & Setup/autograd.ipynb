```python
import torch

# Input value is fixed.
# We are not learning x, so requires_grad is not needed.
x = torch.tensor(2.0)

# w is the learnable parameter.
# requires_grad=True tells PyTorch to track operations on w
# so that we can calculate the gradient of loss with respect to w.
w = torch.tensor(3.0, requires_grad=True)

# This is the actual correct output we want the model to predict.
y_true = torch.tensor(10.0)

# Learning rate controls how big the update step should be.
lr = 0.1

# Forward pass:
# The model prediction is y_pred = w * x.
y_pred = w * x

# Calculate loss using squared error.
# This tells us how far the prediction is from the correct answer.
loss = (y_pred - y_true) ** 2

# Backward pass:
# Computes the gradient of loss with respect to w.
# The result is stored in w.grad.
loss.backward()

print("Before Update : ")
print("w : ", w.item())
print("loss : ", loss.item())
print("gradient : ", w.grad.item())

# Update the weight manually.
# torch.no_grad() is used because we do not want PyTorch
# to track this update operation in the computation graph.
with torch.no_grad():
    # Gradient descent formula:
    # new_w = old_w - learning_rate * gradient
    w -= lr * w.grad

# Important:
# Gradients accumulate in PyTorch by default.
# So we reset the gradient before the next training step.
w.grad.zero_()

print("After Update")
print("w : ", w.item())
```

```python
import torch

# Input data.
# These are the values given to the model.
x = torch.tensor([1., 2., 3., 4.])

# Correct output values.
# The model should learn that y = 2 * x.
y_true = torch.tensor([2., 4., 6., 8.])

# Learnable parameter.
# We start with w = 0, and PyTorch will calculate gradients for it.
w = torch.tensor(0.0, requires_grad=True)

# Learning rate controls how big each update step is.
lr = 0.1

# Number of times we train over the same data.
epochs = 10

for epoch in range(epochs):

    # Step 1: Forward pass / prediction
    # The model predicts output using y_pred = w * x.
    y_pred = w * x

    # Step 2: Calculate loss
    # Mean Squared Error:
    # Measures average squared difference between prediction and true value.
    loss = ((y_pred - y_true) ** 2).mean()

    # Step 3: Backward pass
    # Calculates gradient of loss with respect to w.
    # The gradient is stored in w.grad.
    loss.backward()

    # Step 4: Update weight manually
    # We use torch.no_grad() because updating w should not be tracked
    # in PyTorch's computation graph.
    with torch.no_grad():
        # Gradient descent formula:
        # new_w = old_w - learning_rate * gradient
        w -= lr * w.grad

    # Step 5: Reset gradient
    # PyTorch accumulates gradients by default,
    # so we must clear it before the next epoch.
    w.grad.zero_()

    # Print current training progress
    print(f"Epoch {epoch + 1}: W={w.item():.4f}, Loss={loss.item():.4f}")
```
